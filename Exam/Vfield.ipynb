```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig = plt.figure(figsize=(10,5))
ax = plt.axes(xlim=(-5, 5), ylim=(-5, 5))

def update(i):
    ax.clear()
    data = np.loadtxt(f'avg_vel_{i}.dat')
    x, y = data[:, 0], data[:, 1]
    u, v = data[:, 2], data[:, 3]
    ax.quiver(y, x, u, -v, scale=8)
    ax.set_title('2D Vector Field')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_aspect('equal')

ani = FuncAnimation(fig, update, frames=range(1,101), interval=50)
ani.save("uniform_50_100_8.gif")
plt.show()
```
