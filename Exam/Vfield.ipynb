```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the file
data = np.loadtxt('avg_velocity.dat')

# Extract the coordinates and components of the velocity
x = data[:, 0]
y = data[:, 1]
u = data[:, 2]
v = data[:, 3]

# Plot the vector field
plt.figure(figsize=(10,10))
plt.quiver(y, x, u, v, scale=10)

# Set plot title and labels
plt.title('2D Vector Field')
plt.xlabel('x')
plt.ylabel('y')

# Adjust the aspect ratio
plt.gca().set_aspect('equal')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig = plt.figure(figsize=(10,10))
ax = plt.axes(xlim=(-5, 5), ylim=(-5, 5))

def update(i):
    ax.clear()
    data = np.loadtxt(f'avg_vel_{i}.dat')
    x = data[:, 0]
    y = data[:, 1]
    u = data[:, 2]
    v = data[:, 3]
    ax.quiver(y, x, u, v, scale=15)
    ax.set_title('2D Vector Field')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_aspect('equal')

ani = FuncAnimation(fig, update, frames=range(1,31), interval=300)
ani.save("vfield.gif")
plt.show()
```
